# Splitting data into train and test

```python
x=df.drop('Class',axis=1)
y=df['Class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# Building Model

```python
acc=[]
feature=[]
def model(classifier,df2):    
    x=df2.drop('Class',axis=1)
    y=df2['Class']
    
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=np.random.randint(1,100,1)[0])
    
    feature=x.shape[1]
    
    clf=classifier()
    clf.fit(x_train,y_train)
    pred=clf.predict(x_test)
    
    print('accuracy:',accuracy_score(y_test,pred))
    print('\n')
    print('classification_report:',classification_report(y_test,pred))
    print('confusion_matrix:',confusion_matrix(y_test,pred))
      
```

```python
model(LogisticRegression,df)
```

```python
model(DecisionTreeClassifier,df)
```

```python
model(RandomForestClassifier,df)
```

```python
model(GradientBoostingClassifier,df)
```

# Balancing dataset

```python
sm=SMOTETomek()
```

```python
x_train1,y_train1=sm.fit_resample(x_train,y_train)
```

```python
y_train1.value_counts()
```

# Model result after balancing data

# Logistic Regression

```python
clf=LogisticRegression()
clf.fit(x_train1,y_train1)
pred=clf.predict(x_test)
    
print('accuracy:',accuracy_score(y_test,pred))
print('\n')
print('classification_report:',classification_report(y_test,pred))
print('confusion_matrix:',confusion_matrix(y_test,pred))
```

# DecisionTree

```python
clf=DecisionTreeClassifier()
clf.fit(x_train1,y_train1)
pred=clf.predict(x_test)
    
print('accuracy:',accuracy_score(y_test,pred))
print('\n')
print('classification_report:',classification_report(y_test,pred))
print('confusion_matrix:',confusion_matrix(y_test,pred))
```

# RandomForestClassifier

```python
clf=RandomForestClassifier()
clf.fit(x_train1,y_train1)
pred=clf.predict(x_test)
    
print('accuracy:',accuracy_score(y_test,pred))
print('\n')
print('classification_report:',classification_report(y_test,pred))
print('confusion_matrix:',confusion_matrix(y_test,pred))
```

# GradientBoostingClassifier

```python
clf=GradientBoostingClassifier()
clf.fit(x_train1,y_train1)
pred=clf.predict(x_test)
    
print('accuracy:',accuracy_score(y_test,pred))
print('\n')
print('classification_report:',classification_report(y_test,pred))
print('confusion_matrix:',confusion_matrix(y_test,pred))
```
